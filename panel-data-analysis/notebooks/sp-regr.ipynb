# PySAL

```python
import geopandas as gpd
from libpysal.weights import Rook
from spreg import OLS, ML_Lag, ML_Error, GM_Lag, GM_Error_Het
```

```python
gdf = gpd.read_file('election.shp')
```

```python
y = gdf['pct_gop_16']
X = gdf[['RHI125214', 'EDU685213']]
```

```python
wr = Rook.from_dataframe(gdf, geom_col='geometry')
```

## OLS

Без пространственного лага регрессоров и индексами Морана

```python
ols = OLS(y, X, w=wr, name_y='pct_gop_16' , name_x=['RHI125214', 'EDU685213'], robust='white', moran=True, spat_diag=True)
print(ols.summary)
```

с пространственным лагом регрессора и индексами Морана

```python
ols = OLS(y, X, w=wr, name_y='pct_gop_16' , name_x=['RHI125214', 'EDU685213'], robust='white', moran=True, spat_diag=True, slx_lags=1)
print(ols.summary)
```

## ML

```python
mllag = ML_Lag(y, X, w=wr, name_y='pct_gop_16' , name_x=['RHI125214', 'EDU685213'])
print(mllag.summary)
```

```python
mlerr = ML_Error(y, X, w=wr, name_y='pct_gop_16' , name_x=['RHI125214', 'EDU685213'])
print(mlerr.summary)
```

## GMM

```python
gmlag = GM_Lag(y, X, w=wr, name_y='pct_gop_16' , name_x=['RHI125214', 'EDU685213'], robust='white')
print(gmlag.summary)
```

```python
gmerr = GM_Error_Het(y, X, w=wr, name_y='pct_gop_16' , name_x=['RHI125214', 'EDU685213'])
print(gmerr.summary)
```
